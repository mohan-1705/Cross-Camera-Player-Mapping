```python
# Step 1: Install required lib
!pip install -q ultralytics gdown

# Step 2: Download best.pt from Google Drive
!gdown --id 1-5fOSHOSB9UXyP_enOoZNAMScrePVcMD -O best.pt

# Step 3: Download the two soccer videos
!gdown --id 10U3DzW2EnnTdCuJ8R-QVn4szM6lX5ff5 -O broadcast.mp4
!gdown --id 1bXQi4b_ScghZrg-SB5TEz-dyqsgsPoXo -O tacticam.mp4

# Step 4: Run detection and annotate frame-by-frame
import cv2
from ultralytics import YOLO

def detect_and_save(input_video, output_video, model_path="best.pt", conf=0.3):
    model = YOLO(model_path)

    cap = cv2.VideoCapture(input_video)
    if not cap.isOpened():
        print(f"❌ Error: could not open {input_video}")
        return

    fps = cap.get(cv2.CAP_PROP_FPS) or 30
    w = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
    h = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))

    fourcc = cv2.VideoWriter_fourcc(*"mp4v")
    out = cv2.VideoWriter(output_video, fourcc, fps, (w, h))

    frame_count = 0
    while True:
        ret, frame = cap.read()
        if not ret:
            break
        frame_count += 1
        results = model.predict(frame, conf=conf, verbose=False)[0]
        annotated = results.plot()
        out.write(annotated)

    cap.release()
    out.release()
    print(f"✅ Saved annotated video: {output_video} ({frame_count} frames)")

# Process both videos
detect_and_save("broadcast.mp4", "broadcast_output.mp4")
detect_and_save("tacticam.mp4", "tacticam_output.mp4")
```
